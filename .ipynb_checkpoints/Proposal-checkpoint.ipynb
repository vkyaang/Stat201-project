```r
install.packages('janitor')

library(tidyverse)
library(repr)
library(datateachr)
library(digest)
library(infer)
library(gridExtra)
library(cowplot)
library(janitor)
```

```r
crime_data <- read_csv('crimedata_csv_AllNeighbourhoods_AllYears.csv')
head(crime_data)
```

```r
crime_data_clean <- crime_data %>%
    drop_na() %>%
    mutate(TYPE = as.factor(TYPE)) %>%
    select(TYPE, YEAR, NEIGHBOURHOOD) %>%
    filter(YEAR == c(2019, 2020, 2021, 2022)) %>%
    filter(NEIGHBOURHOOD == c('West End', 'Strathcona')) %>%
    filter(TYPE != 'Homicide')

crime_data_2 <- crime_data %>%
    filter(NEIGHBOURHOOD == c('Strathcona', 'West End')) %>%
    select(TYPE, YEAR, NEIGHBOURHOOD) %>%
    group_by(NEIGHBOURHOOD) %>%
    filter(TYPE == 'Theft from Vehicle')

# head(crime_data_2)

head(crime_data_clean)
summary(crime_data_clean)
```

```r
crime_pop <- crime_data_clean %>%
    group_by(NEIGHBOURHOOD, YEAR) %>%
    summarise(n = n())

crime_pop_plot <- crime_pop %>%
    ggplot(aes(x=YEAR, y=n, color = NEIGHBOURHOOD)) +
    geom_line()

crime_pop
crime_pop_plot
```

```r
options(repr.plot.width = 10, repr.plot.height = 5)
crime_type <- crime_data_clean %>%
    group_by(TYPE, YEAR) %>%
    summarise(n=n())

crime_type_plot <- crime_type %>%
    ggplot(aes(x=YEAR, y=n, colour = TYPE)) +
    geom_line()
head(crime_type)
crime_type_plot
```

```r
theft_vehicle <- crime_data_clean %>%
    filter(TYPE == 'Theft from Vehicle') %>% 
    group_by(NEIGHBOURHOOD, YEAR) %>%
    summarise(n=n()) %>%
    pivot_wider(names_from = NEIGHBOURHOOD, values_from = n) %>%
    clean_names() %>%
    mutate(s_prop = strathcona/theft_count, 
            w_prop = west_end/theft_count,
            prop_diff = s_prop - w_prop)

theft_count <- crime_data_clean %>%
    filter(TYPE == 'Theft from Vehicle') %>%
    group_by(TYPE) %>%
    summarise(n=n()) %>%
    pull()
    

theft_vehicle
theft_count
```

```r
# maybe difference between poor and rich with break in

crime_total <- crime_data_clean %>%
    summarise(total = n())
    
westend_total <- crime_data_clean %>%
    filter(NEIGHBOURHOOD == 'West End') %>%
    summarise(westend_total = n())

strathcona_n <- crime_data_clean %>%
    filter(NEIGHBOURHOOD == 'Strathcona') %>%
    filter(TYPE !='Homicide') %>%
    summarise(Strathcona_total = n())

total <- crime_total %>%
    inner_join(westend_total, by = c('TYPE', 'YEAR')) %>%
    inner_join(strathcona_n, by = c('TYPE', 'YEAR')) 


total
```

```r
total_line <- total %>%
    ggplot(aes(x=YEAR, y=total)) +
    geom_line()

total_line
```

```r
prop <- total %>%
    mutate(westend_prop = westend_total/total)

prop
```
